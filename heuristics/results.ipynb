```python
import os

# Separate lists to store extracted data as integers
n_values = []
m_values = []
width_values = []
time_values = []
nbags_values = []

import csv

# Read data from CSV file
with open('output.csv', mode='r') as file:
    csv_reader = csv.reader(file)
    
    # Skip the header row
    next(csv_reader)

    # Process each row in the CSV
    for row in csv_reader:
        n_values.append(int(row[1]))         # Num Vertices
        m_values.append(int(row[2]))            # Num Edges
        width_values.append(int(row[3]))             # Treewidth
        nbags_values.append(int(row[4]))                  # Bags
        time_values.append(int(row[5]))        # Execution Time (milliseconds)

# Displaying the lists
print("Num Vertices:", n_values)
print("Num Edges:", m_values)
print("Treewidth:", width_values)
print("Bags:", nbags_values)
print("Execution Time (milliseconds):", time_values)

```

```python
import numpy as np
import matplotlib.pyplot as plt

# Combine n_values and width_values into a single array for sorting
combined = list(zip(n_values, width_values))

# Sort the combined list based on n_values
sorted_combined = sorted(combined)

# Unzip the sorted list back into two separate lists
sorted_n_values, sorted_width_values = zip(*sorted_combined)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(sorted_n_values, sorted_width_values, marker='o', linestyle='-')
plt.xlabel('Node')
plt.ylabel('Width')
plt.grid(True)
plt.show()
```
